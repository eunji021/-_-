```python
# while 반복문

num = 1

while num <= 5: #조건
    print(num)
    num += 1 #변경 시켜주는 조건문 적어야 함/num = num + 1

print("while문이 종료된 후 num은 %d입니다." % num)
```

```python
#break문

num = 1

while True:
    print(num)
    num +=1
    if num > 5:
        break
```

```python
#continue문

num = 0

while num <= 10:

    num += 1

    if num <5:
       continue
    print(num)
```
